# Bayg29

actually first of all we have to input bayg29.txt with panda library
after that we set all parameters we need for this test case
1. cities : equal number of vector that exist in this test case
2. SP : Size of first population and all population we create in NextGeneration()
3. TT : equal number of player attend in Tournament Selection
4. PC : equal probability of Recombination for 2 parents
5. PM : equal probability of Mutation for one child
6. MIteration : number of generation we want to pass

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import rand,randint
```

```python
inputBayg=pd.read_csv("bayg29.txt",header=None,delim_whitespace=True)
```

```python
cities=29
SP=20
TT=3
PC=0.9
PM=0.05
MIteration=1000
PR=0.3
```

```python
distances=np.zeros((cities,cities))
```

```python
for i in range(cities):
    for j in range(i):
        distances[i][j]=inputBayg[i-j-1][j]
        distances[j][i]=distances[i][j]
pd.DataFrame(distances)
```

# Fitness Function

it is obvious that fitness for any circle is 1/ lenPath 
lenPath is sum of all distances that we are pass in this circle
because of we want to produce faster and we don't see any problem in float number we use lenPath 
and in Comparison of two circle we say if on circle have a short lenPath than other we say it has better Fitness

```python
def CalcLength(circle):
    lenPath=0
    for i in range(1,cities):
        lenPath+=distances[int(circle[i])][int(circle[i-1])]
    lenPath+=distances[int(circle[0])][int(circle[cities-1])]
    return lenPath
```

```python
def PutLengthInCol(circle):
    return np.array((CalcLength(circle),circle),dtype=object)
```

```python
def CreateRandomTour():
    circle=np.arange(cities)
    np.random.shuffle(circle)
    return PutLengthInCol(circle)
```

```python
CreateRandomTour()
```

# First Generation

we use numpy library to shuffle an ordered sequence 

```python
def CreateFirstGeneration():
    return np.array([CreateRandomTour() for _ in range(SP)])
```

```python
pd.DataFrame(CreateFirstGeneration())
```

# Tournament Selection
we use Tournament selection for select who one be parent 

```python
def Selection(pop):
    best=randint(len(pop))
    for kal in randint(0,len(pop)-1,TT-1):
        if pop[best][0]>pop[kal][0]:
            best=kal
    return pop[best]
```

```python
def KSelection(pop,t):
    return np.array([Selection(pop) for _ in range(t)])
```

# Crossover
for this we use ordered recombination and select two point in parent 
and use ordered recombination algorithm to create two child from them

```python
def HelpCrossover(parent1,parent2,child,x,y):
    child[0:x]=-1
    child[y:len(parent1)]=-1
    if y==len(parent1):
        y=0
    d=y
    while d!=x:
        if parent2[y] not in child:
            child[d]=parent2[y]
            d+=1
            if d==len(parent1):
                d=0
        y+=1
        if y==len(parent1):
            y=0
    if -1 in child:
        display("ah",x,y,pd.DataFrame(child))
```

```python
def GetXYRandom(size):
    x,y=randint(0,size,2)
    y+=1
    while y==x:
        y=randint(0,size)
    if y<x:
        x,y=y,x
    
    return (x,y)
```

```python
def Crossover(parent1,parent2):
    child1=parent1.copy()
    child2=parent2.copy()
    if rand() < PC:
        x,y=GetXYRandom(len(parent1))
        HelpCrossover(parent1,parent2,child1,x,y)
        HelpCrossover(parent2,parent1,child2,x,y)
    if -1 in child1:
        display("slm",pd.DataFrame(child1))
    return [child1,child2]
```

# Mutation
we use Scramble mutation and as we say befor , we use PM for probability of mutation

```python
def Mutation(child):
    if rand()< PM:
        x,y=GetXYRandom(len(child))
        np.random.shuffle(child[x:y])
```

```python
def CreateChildren(pop):
    parents=KSelection(pop,len(pop))
    children=np.empty((len(pop),2),dtype=object)
    for i in range(0,len(parents),2):
        glay=Crossover(parents[i][1],parents[i+1][1])
        for j in range(2):
            children[i+j][1]=glay[j]
            Mutation(children[i+j][1]) 
            children[i+j]=PutLengthInCol(children[i+j][1])
    return children
```

# Find Best And Average
Here we user numpy.array (x,y,z)
1. x : equal average of the generatation
2. y : equal length of best circle
3. z : exact best circle of the generation

```python
def FindETPop(pop):
    ans=np.zeros(len(pop[0])+1,dtype=object)
    t=0
    ans[0]=pop[0][0]
    for i in range(1,len(pop)):
        if pop[i][0]<pop[t][0]:
            t=i
        ans[0]+=pop[i][0]
    ans[0]/=len(pop)
    ans[1:]=pop[t]
    return ans
```

```python
def CreateNextGeneration(pop):
    children=CreateChildren(pop)
    pop[1]=FindETPop(children)[1:].copy()
    pop[0]=FindETPop(pop)[1:].copy()
    pop[int(len(pop)*PR):]=children[len(pop)-int(len(pop)*PR)]
```

# analysis 
we have to say in this test case if we use better selection for replacement 
we have better circle but i think it is enough good
red chart stands for **Average** and blue on stands for **Best Length**

```python
def Genetic_Algorithm():
    pop=CreateFirstGeneration()
    BestAnswer=np.empty((MIteration,3),dtype=object)
    BestAnswer[0]=FindETPop(pop)
    for i in range(MIteration-1):
        CreateNextGeneration(pop)
        BestAnswer[i+1]=FindETPop(pop)
    display(pd.DataFrame(BestAnswer[:]))
    plt.plot(BestAnswer[:,1],'b')
    plt.plot(BestAnswer[:,0],'r')
```
